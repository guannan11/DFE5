```python
# 1.import libraries
import pandas as pd
```

```python
# 3.assign it to baby_name
baby_names = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')
baby_names.info()
```

```python
# 4. show the first 10 entries
baby_names.head(10)
```

```python
# 5. delete 'Id' & 'Unnamed: 0' column
del baby_names['Unnamed: 0']
del baby_names['Id']
baby_names.head()
```

```python
# 6. number of male and female
baby_names['Gender'].value_counts()
```

```python
## 7. group the dataset by name, assign them to names
# delect 'Year' column as you don't want to sum it
del baby_names["Year"]
# group the data by name
names = baby_names.groupby("Name").sum()
# print the fist 5 entries
names.head()
# print the size of the dataset
print(names.shape)
# sort it in descending order
names.sort_values("Count", ascending = 0).head()
```

```python
# 8.how many different names in dataset (as we have already 
#   grouped by the name, all the names are unique already)
len(names)
```

```python
# 9. which name occur most
names.Count.idxmax()
# or
# names[names.Count == names.Count.max()]
```

```python
# 10. how many names has occured least
len(names[names.Count == names.Count.min()]) 
```

```python
# 11. name occurrence median
names[names.Count == names.Count.median()]
```

```python
# 12. standard deviation of names
names.Count.std()
```

```python
# 13. summary with the mean, min, max, std and quartiles
names.describe()
```
