```python
# versus-> MATRIX-> transpose
# T-> converts rows into columns and vice versa
```

```python
stats = data.describe().T
stats.columns
```

```python
# taking transpose helps in directly selecting the name of columns we need!

allmeans = stats['mean']
allstd = stats['std']
```

```python
z_score = (data-allmeans)/allstd
z_score
# -inf to +inf where the mean is assumed to be 0, and all other values
# have been scaled with respect to the mean
```
