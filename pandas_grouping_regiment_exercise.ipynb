```python
# 1. import libraries
import pandas as pd
```

```python
# 2. create the DataFrame
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'], 
        'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'], 
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'], 
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
```

```python
# 3. assgin it to regiment
regiment = pd.DataFrame(raw_data, columns = raw_data.keys())
regiment
```

```python
# 4. mean pre test score for Nighthawks
regiment[regiment['regiment'] == 'Nighthawks'].groupby('regiment').mean()
```

```python
# 5. general statistics by company
regiment.groupby('company').describe()
```

```python
# 6.mean of each company's preTestScore
regiment.groupby('company').preTestScore.mean()
```

```python
# 7. mean preTestScores grouped by regiment and company
regiment.groupby(['regiment', 'company']).preTestScore.mean()
```

```python
# 8.mean preTestScores grouped by regiment and company without heirarchical indexing
regiment.groupby(['regiment', 'company']).preTestScore.mean().unstack()
```

```python
# 9. Group the entire dataframe by regiment and company
regiment.groupby(['regiment', 'company']).mean()
```

```python
# 10. number of observations in each regiment and company
regiment.groupby(['regiment', 'company']).size()
```

```python
# 11. seperate the entire data by regiments
for name, group in regiment.groupby('regiment'):
    print(name)
    print(group)
```
