```python
# 1. import libraries
import pandas as pd
```

```python
# 2. data given as dictionary
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
            'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
            'deaths': [523, 52, 25, 616, 43, 234, 523, 62, 62, 73, 37, 35],
            'battles': [5, 42, 2, 2, 4, 7, 8, 3, 4, 7, 8, 9],
            'size': [1045, 957, 1099, 1400, 1592, 1006, 987, 849, 973, 1005, 1099, 1523],
            'veterans': [1, 5, 62, 26, 73, 37, 949, 48, 48, 435, 63, 345],
            'readiness': [1, 2, 3, 3, 2, 1, 2, 3, 2, 1, 2, 3],
            'armored': [1, 0, 1, 1, 0, 1, 0, 1, 0, 0, 1, 1],
            'deserters': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
            'origin': ['Arizona', 'California', 'Texas', 'Florida', 'Maine', 'Iowa', 'Alaska', 'Washington', 'Oregon', 'Wyoming', 'Louisana', 'Georgia']}
```

```python
# 3. create a dataframe, assign it to army (variable)
army = pd.DataFrame(data=raw_data)
army
```

```python
# 4. set the 'origin' column as the index of the dataframe
army.set_index('origin', inplace=True)
```

```python
# 5. print 'veterans' column
army.veterans
```

```python
# 6. print 'veterans' & 'deaths' columns
army[["veterans", "deaths"]]
```

```python
# 7. print the name of all the columns
army.columns
```

```python
# 8. Select the 'deaths', 'size' and 'deserters' columns from Maine and Alaska
army.loc[["Maine", "Alaska"], ["deaths", "size", "deserters"]]
```

```python
# 9. Select the rows 3 to 7 and the columns 3 to 6
army.iloc[2:7, 2:6]
```

```python
# 10. select all rows after 4th row, and all columns
army.iloc[4:, :]
```

```python
# 11. select all row up to the 4th, and all columns
army.iloc[:4, :]
```

```python
# 12. select from 3rd to 7th columns
army.iloc[:, 2:7]
```

```python
# 13. Select rows where df.deaths is greater than 50
army[army["deaths"] > 50]
```

```python
# 14. Select rows where df.deaths > 500 or < 50
army[(army["deaths"] > 500) | (army["deaths"] < 50)]
```

```python
# 15. Select all the regiments not named "Dragoons"
army[army["regiment"] != "Dragoons"]
```

```python
# 16. select all rows called Texas and Arizona
army.loc[["Texas", "Arizona"], :]
```

```python
# 17. select 3rd cell in row Arizona
army.loc[["Arizona"]].iloc[:, 2]
```

```python
# 18. select 3th cell in deaths columns
army.loc[:, ["deaths"]].iloc[2]
```
