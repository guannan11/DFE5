```python
# 1. import libraries
import pandas as pd
import numpy as np
```

```python
# 2. assign them to cars1 & cars2
cars1 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv")
cars2 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv")

print(cars1.head())
print(cars2.head())
```

```python
# 4. first dataset has some unnamed blank columns, fix cars1
cars1 = cars1.loc[:, "mpg":"car"]
cars1.head()
```

```python
# 5. the dimensions in each datasets
print(cars1.shape)
print(cars2.shape)
```

```python
# 6. join 2 datasets into one DataFrame named cars
cars = cars1.append(cars2)
cars
```

```python
# 7. owner column is missing, Create a random number Series from 15,000 to 73,000.
nr_owners = np.random.randint(15000, high=73001, size=398, dtype='l')
nr_owners
```

```python
# 8. add owners column in cars dataframe
cars['owners'] = nr_owners
cars.tail()
```
