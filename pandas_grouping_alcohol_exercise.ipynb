```python
# 1. import libraries
import pandas as pd
```

```python
# 3. assign it to drinks (variable)
drinks = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
drinks.head()
```

```python
# 4. which continent drinks more beer on average
drinks.groupby('continent').beer_servings.mean()
```

```python
# 5. print the statistics for wine consumption for each continent
drinks.groupby('continent').wine_servings.describe()
```

```python
# 6. print the mean of every drinks per continent
drinks.groupby('continent').mean()
```

```python
# 7. print the median of every drinks per continent
drinks.groupby('continent').median()
```

```python
# 8.Print the mean, min and max for spirit per continent
drinks.groupby('continent').spirit_servings.agg(['mean', 'min', 'max'])
```
