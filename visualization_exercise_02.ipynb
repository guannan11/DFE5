```python
# 1. import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
sns.set_style("white")
```

```python
# 3. assign it to tips (variable)
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv'
tips = pd.read_csv(url)
tips.head()
```

```python
# 4. delete the Unnamed 0 column
del tips['Unnamed: 0']
tips.head()
```

```python
# 5. plot the total_bill column histogram
ttbill = sns.distplot(tips.total_bill);
ttbill.set(xlabel ='Value', ylabel ='Frequency', title = "Total Bill")
sns.despine()
```

```python
# 6. scatter plot show the relationship between total_bill and tip
sns.jointplot(x="total_bill", y ="tip", data = tips)
```

```python
# 7. image shows the relationship of total_bill, tip and size
sns.pairplot(tips)
```

```python
# 8. show relationship between days and total_bill
sns.stripplot(x = "day", y = "total_bill", data = tips, jitter = True);
```

```python
# 9. scatter plot, x-tip, y-day, differ the dots by sex
sns.stripplot(x = "tip", y = "day", hue = "sex", data = tips, jitter = True);
```

```python
# 10. box plot to show total_bill per day at Dinner and lunch 
sns.boxplot(x = "day", y = "total_bill", hue = "time", data = tips);
```

```python
# 11. 2 side by side histograms (dinner vs tip, lunch vs tip)
sns.set(style = "ticks")
g = sns.FacetGrid(tips, col = "time")
g.map(plt.hist, "tip");
```

```python
# 12. 2 side by side scatterplots graphs, one for male, one for female, 
#     tip vs total_bill, differ by smoker and no smoker 
g = sns.FacetGrid(tips, col = "sex", hue = "smoker")
g.map(plt.scatter,"total_bill", "tip", alpha =.7)
g.add_legend;
```
