```python
# 1. Import libraries
import pandas as pd
import datetime

# 3. assign it to data, replace the first 3 columns with a proper datetime index
data_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data'
data = pd.read_csv(data_url, sep = "\s+", parse_dates = [[0,1,2]]) 
data.head()
```

```python
## 4. 2061 doesn't seem right, create a function to fix it and apply it.
# function that uses datetime
def fix_century(x):
  year = x.year - 100 if x.year > 1989 else x.year
  return datetime.date(year, x.month, x.day)
# apply the function fix_century on the column and replace the correct values
data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_century)
data.head()
```

```python
## 5. set dates as index (data type is datetime64[ns])
# transform Yr_Mo_Dy to date type datetime64
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])
data = data.set_index('Yr_Mo_Dy')
data.head()
```

```python
# 6. how many values are missing in each cloumn
data.isnull().sum()
```

```python
# 7. how many non-missing values in each column
data.shape[0] - data.isnull().sum()
# or
data.notnull().sum()
```

```python
# 8. mean windspeeds of the windspeeds over all the locations and all the times.
data.sum().sum() / data.notna().sum().sum()
```

```python
## 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds
##     and standard deviations of the windspeeds at each location over all the days
data.describe(percentiles=[])
# by default not setting this would have shown you Inter-quartile ranges (25% and 75% values)
```

```python
## 10. Create a DataFrame called day_stats and calculate the min, max, mean windspeed
##      and standard deviations of the windspeeds across all the locations at each day.
#  create the dataframe
day_stats = pd.DataFrame() # determine axis equals to 1 so it gets each row
# determine axis equals to one so it gets each row
day_stats['min'] = data.min(axis = 1) # min
day_stats['max'] = data.max(axis = 1) # max 
day_stats['mean'] = data.mean(axis = 1) # mean
day_stats['std'] = data.std(axis = 1) # standard deviations

day_stats.head()
```

```python
# 11. Find the average windspeed in January for each location
data.loc[data.index.month == 1].mean()
```

```python
# 12. Downsample to yearly record for each location.
data.groupby(data.index.to_period('A')).mean()
```

```python
# 13. downsample to a monthly record for each location
data.groupby(data.index.to_period('M')).mean()

# 14. downsample to a weekly record for each location
data.groupby(data.index.to_period('W')).mean()
```

```python
## 15. min, max, mean, standard deviations of the windspeeds for all 
## locations for the first 52 weeks.(1st week starts on 1961/Jan/02)
# resample data to 'W' week and use the functions
weekly = data.resample('W').agg(['min', 'max', 'mean','std'])
# slice it for the first 52 weeks and locations
weekly.loc[weekly.index[1:53], "RPT":"MAL"] .head(10)
```
